# Data Cleaning

* Common theme of doing things in parallel.
* Have your data in flat dataframe, sorted by ticker then date.
* To do operations, use groupby > apply > lambda > assign paradigm.
* Grouping by date is cross-sectional.
* Grouping by asset is time-based.

## Load Data

```python
import pandas as pd
import requests
import numpy as np
import matplotlib.pyplot as plt
import warnings
import seaborn as sns
sns.set_style('whitegrid')
warnings.simplefilter(action='ignore', category=FutureWarning)


perps = pd.read_csv('binance_perp_daily.csv')
```

## Remove Stablecoins

```python
url = "https://stablecoins.llama.fi/stablecoins?includePrices=true"
response = requests.get(url)
data = response.json()
stables = pd.Series([asset["symbol"] + 'USDT' for asset in data["peggedAssets"]]+['FTTUSDT'])
perps = perps[~perps['ticker'].str.match('|'.join([f'{s}' for s in stables]))]
```

## Coins by Volume

* First groupby ticker, then calculate trailing volume.
* Then groupby date, rank it, filter by rank.
* Again, groupby-apply-lambda-assign.
* Dynamic universe selection at each point-in-time with a flat dataframe.

```python
trading_universe_size = 30

# Group by ticker, calculate trailing 30-day volume, and filter out NA values
universe = perps.groupby('ticker') \
    .apply(lambda x: x.assign(trail_volume=x['dollar_volume'].rolling(window=30).mean())) \
    .dropna() \
    .groupby('date') \
    .apply(lambda x: x.assign(volume_rank=x['trail_volume'].rank(ascending=False),
                              is_universe=x['trail_volume'].rank(ascending=False) <= trading_universe_size))

# Group by date and is_universe, summarize count, and reset index
universe_counts = universe.groupby(['date', 'is_universe']).size().reset_index(name='count')

# Plot
plt.figure(figsize=(7, 3.5))
plt.plot(universe_counts[universe_counts['is_universe']]['date'], 
         universe_counts[universe_counts['is_universe']]['count'], 
         label='In Universe', color='blue')
plt.plot(universe_counts[~universe_counts['is_universe']]['date'], 
         universe_counts[~universe_counts['is_universe']]['count'], 
         label='Outside Universe', color='red')
plt.title('Universe Size')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.xticks([])  # Hide x-axis ticks
plt.show()

original_universe = universe.copy()
# reset our index to be sorted by ticker then date and filter our universe to keep only our top 30
universe = universe[universe.is_universe==True].sort_values(['ticker','date']).reset_index(drop=True)
```

## Returns

```python
universe = universe.groupby('ticker')        \
                   .apply(lambda x:x         \
                   .assign(                  \
                    rets = (x['funding_rate'] + (universe['close'] - x['close'].shift()) / x['close'].shift()).shift(-1)
                   ))                        \
                   .dropna()

universe['log_rets'] = np.log(1+universe.rets) 
                        
```

# Features


## Feature Creation

```python

features = universe.groupby('ticker') \
    .apply(lambda x: x.assign( \
        breakout=(9.5 - x['close'].rolling(20).apply(lambda x: x.index.values.max()-x.idxmax())).shift(),
        momo=x['close'].sub(x['close'].shift(10)).div(x['close'].shift(10)).shift(),
        carry=x['funding_rate'].shift()
        )) \
    .reset_index(drop=True) \
    .dropna()

```

## Feature Standardization



```python
scaled_features = features.groupby('date').apply(lambda x:x.assign(
                  demeaned_rets = x.log_rets - x.log_rets.mean(),
                  z_carry = (x.carry - x.carry.mean())/x.carry.std(),
                  decile_carry = pd.qcut(x.carry.rank(method='first'), q=10, labels=False, duplicates='drop') + 1,
                  z_momo = (x.momo - x.momo.mean())/x.momo.std(),
                  decile_momo = pd.qcut(x.momo.rank(method='first'), q=10, labels=False, duplicates='drop') + 1,
                  z_breakout = (x.breakout - x.breakout.mean())/x.breakout.std()
)).reset_index(drop=True)

# Drop rows with NaN values
scaled_features = scaled_features.dropna()


```

```python
# Plot using Seaborn
fig, ax = plt.subplots(1,3, figsize=(13,3))
ax.flatten()

# Create separate subplots for each feature type
for i, feature in enumerate(['z_breakout', 'z_momo', 'z_carry']):
    scaled_features[feature].plot(kind='hist', bins=200, ax=ax[i], title=f'{feature}')

plt.show()
```

## Signal IC

* Pandas
    * For pandas, effective use of `melt` to long form, then `groupby`-`apply`-`lambda`.
    * For `melt`,`id_vars` remain unchanged, `value_vars` will be the column names melted into single column. Their `value_name` is the singular column of their values.
    * Once done, easy to `groupby`. `apply` a `lambda` function to return a series, for each group of correlation.
* Features
    * Carry is clearly the strongest predictive feature of forward demeaned returns.
    * Momo shows the slight mean reversion effect.

```python


# Assuming 'features_scaled' DataFrame is already defined

# Melt the DataFrame to long format
melted_features = pd.melt(scaled_features, id_vars=['date', 'ticker', 'is_universe', 'demeaned_rets'], 
                           value_vars=['breakout', 'z_carry', 'z_momo', 'decile_carry', 'decile_momo'], 
                           var_name='feature', value_name='value')

# Calculate Information Coefficient (IC)
ic_data = melted_features.groupby('feature').apply(lambda x: pd.Series({'IC': x['value'].corr(x['demeaned_rets'])})).reset_index()


# Plot using Seaborn
plt.figure(figsize=(9, 3))
sns.barplot(x='feature', y='IC', data=ic_data, order=['breakout', 'z_carry', 'decile_carry', 'z_momo', 'decile_momo'], palette='viridis')

# Set labels and title
plt.xlabel('Feature')
plt.ylabel('IC')
plt.title('Relative Return Information Coefficient')

plt.show()
```

## Signal Decay (Forward IC)

* Pandas
    * Same workflow as above, `melt` with `groupby` but a double `melt`, then a `lineplot` with `hue` by `feature`.
* Features
    * Original plot has sticky IC for carry (but lower starting point), slower decay. Maybe because of my universe selection.
    * Momo starts with negative IC.
    * Breakout has very sticky IC in time series return.

```python
forward_periods = range(2, 8)
for i in forward_periods:
    scaled_features[f'demeaned_rets_{i}'] = scaled_features.groupby('ticker')['demeaned_rets'].shift(-i)

# Melt the DataFrame to long format
melted_features = pd.melt(scaled_features, id_vars=['date', 'ticker', 'is_universe', 'demeaned_rets'] + [f'demeaned_rets_{i}' for i in range(2,8)], 
                           value_vars=['z_carry', 'z_momo', 'decile_carry', 'decile_momo'], 
                           var_name='feature', value_name='value')

melted_melted_features = pd.melt(melted_features, id_vars=['feature','value'],
                          value_vars=['demeaned_rets']+[f'demeaned_rets_{i}' for i in range(2,8)],
                          var_name='lookahead',value_name='rets')

ic_data = melted_melted_features.groupby(['feature', 'lookahead']).apply(
    lambda x: pd.Series({'IC': x['value'].corr(x['rets'])})
).reset_index()
```

```python
fig, ax = plt.subplots(figsize=(10,4))
sns.lineplot(data=ic_data, x='lookahead', y='IC', hue='feature', style='feature', markers=True, ax=ax, markersize=8, palette='viridis')
plt.title('IC by forward period (days) against relative returns')
plt.xlabel('Forward Period')
plt.show()
```

# Combining Signals

* Pandas
    * `transform` vs `apply`: they do the same thing, but the former will _broadcast_ result, the latter will _reduce_.
    * Instead of `groupby`-`apply`-`lambda`-`assign`, we can `assign`-`transform`.
    * The procedure takes deciles (1-10), and centers them about 0.
* Weights
    * Overweights carry (high IC), negatively correlated with momo/breakout. 
    * Underweight momo (low IC)
    * Overweight breakout slightly. Since it is weighted by time-series not cross-section, there is a net tilt effect (E.g if high breakouts for all the perps then will net long)
* Returns
    * Not a backtest, it says plots returns to combined features if you could do so frictionlessly.
* Lookahead

## Model

```python

min_trading_universe_size = 3

# Find the start date based on the minimum trading universe size
start_date = (
    scaled_features.groupby(['date', 'is_universe'])
    .size()
    .reset_index(name='count')
    .query('count >= @min_trading_universe_size')
    .sort_values('date')
    .iloc[0]['date']
)

# Filter and preprocess the DataFrame for modeling using assign
model_df = (
    scaled_features\
    [['ticker','date','close','funding_rate','z_carry','momo','log_rets','demeaned_rets','is_universe','decile_carry','decile_momo','breakout']]
    .query('is_universe and date >= @start_date')
    .assign(
        carry_weight=lambda x: x['decile_carry'] - 5.5,
        momo_weight=lambda x: (x['decile_momo'] - 5.5),
        breakout_weight=lambda x: x['breakout'] / 2,
        combined_weight=lambda x: 0.5 * x['carry_weight'] + 0.2 * x['momo_weight'] + 0.3 * x['breakout_weight'])  
    ) \
    .groupby('date').apply(lambda x:x.assign( scaled_weight = np.where(x.combined_weight==0,0,x.combined_weight/x.combined_weight.abs().sum()))) \
    .groupby('date').apply(lambda x:x.assign(  
                                            scaled_carry_weight = np.where(x.carry_weight==0,0,x.carry_weight/x.combined_weight.abs().sum()),
                                            scaled_momo_weight = np.where(x.momo_weight==0,0,x.momo_weight/x.combined_weight.abs().sum()),
                                            scaled_breakout_weight = np.where(x.breakout_weight==0,0,x.breakout_weight/x.combined_weight.abs().sum())
                                             ))\


```

```python
model_df['carry_rets'] = model_df.log_rets* model_df.scaled_carry_weight
model_df['momo_rets'] = model_df.log_rets* model_df.scaled_momo_weight
model_df['breakout_rets'] = model_df.log_rets* model_df.scaled_breakout_weight

model_df.groupby('date')['carry_rets','momo_rets','breakout_rets'].sum().cumsum().plot(figsize=(12,4))

plt.title('Carry, Momentum, Breakout factor returns on top 30 perps by volume')
plt.xlabel('')
plt.ylabel('Cumulative return')
```

## Rolling Regressions


* End result: a dataframe of regression coefficients, daily row, 2 columns (momo, carry).
* Dataframe of coefficients, row is every 10 days (e.g 10 Jan,
 20 Jan, 30 Jan) with coeffs. Resample and backfill. Join to model_df.
* Iterate, calculate coefficients, append to df.
* Some more work to convert the new data (weights and coefficients) into _expected returns_
* Wrap backtest into Python script.

```python
df = model_df.copy()
df = df.groupby('date').apply(lambda x:x.assign(
                  breakout_weight = pd.qcut(x.breakout_weight.rank(method='first'), q=10, labels=False, duplicates='drop') + 1 - 5.5,
)).reset_index(drop=True)
```

```python

import statsmodels.api as sm


def rolling_ols(data, target_col, feature_cols, window_size=30, skip_size=5):
    
    label = 'xs' if target_col == 'demeaned_rets' else 'ts'
    results = pd.DataFrame(columns=['date'] + [f'coef_{col}' for col in feature_cols] + [f'{label}_intercept'] + [f'{label}_r2'])

    # Group the data by 'date'
    grouped_data = data.groupby('date')
    # Iterate through the unique dates with the rolling window and skip
    for i in range(0, len(grouped_data), skip_size):
        # Select the unique dates for the current window
        selected_dates = list(grouped_data.groups.keys())[i:i+window_size]
        # Extract rows for the selected dates
        window_data = data[data['date'].isin(selected_dates)]

        print(window_data.date.min())
        print(window_data.date.max())
        # Fit the OLS model
        y = window_data[target_col]
        X = sm.add_constant(window_data[feature_cols])
        model = sm.OLS(y, X).fit()
        # Extract and store the results
        results = results.append({'date': window_data.iloc[-1]['date'],
                                        **{f'coef_{col}': model.params[col] for col in feature_cols},
                                        f'{label}_intercept': model.params['const'],
                                        f'{label}_r2': model.rsquared},
                                        ignore_index=True)

    results.set_index('date', inplace=True)
    results.index = pd.to_datetime(results.index)
    return results.groupby(results.index).mean()
```

```python

xs_coeffs = rolling_ols(df, 'demeaned_rets', ['carry_weight','momo_weight','breakout_weight']).resample('D').ffill()
exp_return_df = df 
exp_return_df['date'] = pd.to_datetime(exp_return_df['date'])
exp_return_df = pd.merge(exp_return_df,xs_coeffs,on='date',how='inner')
```

```python
xs_coeffs[['coef_carry_weight','coef_momo_weight','coef_breakout_weight']].plot(figsize=(10,4),title='Rolling linear model coefficients')
```

## Blending Coefficients into Weights

```python


exp_return_df['combined_weight'] = (exp_return_df['carry_weight'] * exp_return_df['coef_carry_weight'] +
                                       exp_return_df['momo_weight'] * exp_return_df['coef_momo_weight'] +
                                       exp_return_df['breakout'] * exp_return_df['coef_breakout_weight']+
                                         exp_return_df['xs_intercept']   )



exp_return_df = exp_return_df.groupby('date').apply(lambda x:x.assign(scaled_weight = np.where(x.combined_weight==0,0,x.combined_weight/x.combined_weight.abs().sum())))
```

```python
exp_return_df['port_rets'] = exp_return_df.log_rets* exp_return_df.scaled_weight
exp_return_df.groupby('date').port_rets.sum().cumsum().plot(figsize=(12,4))

plt.title('Combined Carry, Momentum, and Breakout Model on top 30 perps by volume')
plt.xlabel('')
plt.ylabel('Cumulative return')
```
